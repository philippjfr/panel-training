```python
from collections import Counter
```

```python
PORTS = {i: [6001] for i in range(8)}

def port(worker_id, worker_count):
    n = int(re.sub(r"\D", "", worker_id))
    new_port = PORTS[n][0] + n
    PORTS[n][0] += worker_count
    print(new_port)
    return new_port

c = Counter()
for _ in range(10):
    for i in range(8):
        c[(port(f'gw{i}', 8))] += 1
```

```python
from panel.reactive import ReactiveHTML


class Cookies(ReactiveHTML):

    _cookies = param.String()

    _set = param.String(allow_None=True)

    _scripts = {
        'render': 'data._cookies = document.cookie',
        'set': 'if (data._set) { document.cookie = model._set; }'
    }

    def set_cookie(self, cookie):
        self._set = cookie
        self._set = None

cookie = Cookies()
cookie
```
