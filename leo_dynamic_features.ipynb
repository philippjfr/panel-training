```python
import panel as pn

pn.extension()

n = pn.widgets.IntSlider(start=1, end=20, value=3, name='N')

state = pn.pane.JSON({})

def update_state(event):
    state.object[event.obj.name] = event.new
    state.param.trigger('object')

def create_dynamic_sliders(n):
    sliders = []
    state.object.clear()
    for i in range(n):
        slider = pn.widgets.FloatSlider(start=0, end=7.4, name=f'Feature {i}')
        slider.param.watch(update_state, 'value')
        slider.param.trigger('value')
        sliders.append(slider)
    return sliders

widget_box = pn.WidgetBox(objects=pn.bind(create_dynamic_sliders, n))

pn.Row(
    n,
    widget_box,
    state
)
```
